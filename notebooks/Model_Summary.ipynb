```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = 'Model_Summary.csv'  # Update the path as needed
df = pd.read_csv(file_path)

# Filter the DataFrame for Class 1 recall values
df_class_1 = df[df['Class'] == 1]

# Create the line chart
plt.figure(figsize=(12, 6))
plt.plot(df_class_1['Scenario'], df_class_1['Recall'], marker='o', linestyle='-', color='b')
plt.xlabel('Features')
plt.ylabel('Class 1 Recall')
plt.title('Class 1 Recall vs Features')
plt.xticks(df_class_1['Scenario'])  # Set x-ticks without rotation
plt.grid(True)
plt.tight_layout()

# Display the plot
plt.show()
```

```python
from tabulate import tabulate

# Filter the DataFrame for the required columns
df_scenario_feature = df[['Scenario', 'Features']]

# Generate the table using tabulate
table = tabulate(df_scenario_feature, headers='keys', tablefmt='grid')

# Display the table
print(table)
```
